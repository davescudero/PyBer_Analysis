```python
# To make sure our plots are displayed inline within JupyterNotebook directly below the code cell taht produced it
# The % sign is called "magic command"
%matplotlib inline

# Import dependencies
# pyplot provides a MATLAB-like interface for makeing plots
import matplotlib.pyplot as plt
import statistics
```

```python
# Add some data
# Data is in the form of arrays, or lists. One array will contain the months of the year, and the second will contain the total fares in US dollars for each month

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Line chart using object-oriented interface method
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Another form of doing it 
fig = plt.figure()
ax = fig.add_subplot ()
ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Using MATLAB ceate the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color = "green", linewidth=2, label= "Boston")
# Create labels for the x and y axes
plt.xlabel ("Date")
plt.ylabel ("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Set the x limit between January through June
plt.xlim(0, 5)
# Create a title
plt.title ("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```
